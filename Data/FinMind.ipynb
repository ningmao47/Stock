# 股價日成交資訊 TaiwanStockPrice

{
    date: str, # 日期
    stock_id: str, # 股票代碼
    Trading_Volume: int64, # 成交量
    Trading_money: int64, # 成交金額
    open: float64, # 開盤價
    max: float64, # 最高價
    min: float64, # 最低價
    close: float64, # 收盤價
    spread: float64, # 漲跌幅
    Trading_turnover: float32 # 交易筆數
}

```python
from FinMind.data import DataLoader

api = DataLoader()
api.login_by_token(api_token="")
# api.login_by_token(api_token="你的API_TOKEN")

stock_id="6257"
start_date = "2023-05-01"
end_date = "2026-07-01"

df = api.taiwan_stock_daily(
    stock_id=stock_id,
    start_date=start_date,
    end_date=end_date)
df.head()
```

```python
import pandas as pd

df.to_csv("6257.csv", index=False)
df = pd.read_csv("6257.csv")
df.head()
```

# 法人買賣表 TaiwanStockInstitutionalInvestorsBuySell

{
    date: str, # 日期
    stock_id: str, # 股票代碼
    buy: int64, # 買進
    name: str, # 類別
    sell: int64 # 賣出
}

| name                  | 中文        | 說明                  |
| --------------------- | --------- | ------------------- |
| `Foreign_Investor`    | 外資        | 外資及陸資（主要的外資）        |
| `Foreign_Dealer_Self` | 外資自營商     | 外資券商自己的交易部位         |
| `Investment_Trust`    | 投信        | 國內投信公司              |
| `Dealer_self`         | 自營商（自行買賣） | 證券自營商自行操作的部位        |
| `Dealer_Hedging`      | 自營商（避險）   | 自營商因權證、選擇權等避險需求進行交易 |


```python
stock_id="2330"
start_date = "2026-01-01"
end_date = "2026-07-17"

df = api.taiwan_stock_institutional_investors_wide(
    stock_id=stock_id,
    start_date=start_date,
    end_date=end_date)
df.tail(10)
```
